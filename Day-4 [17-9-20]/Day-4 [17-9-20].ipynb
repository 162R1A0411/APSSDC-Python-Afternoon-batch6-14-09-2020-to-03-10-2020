## Conditional Statments

```python
# if elif
age=15
if age<13:
    print("Not a Teenager")
elif age>18:
    print("Teenager")
else:
    print("Nothing")
```

```python
# Example -2
per=int(input("Enter Per: "))
if (per>=90):
    print("A")
elif per>=80:
    print('B')
elif per>=70:
    print('C')
elif per>=60:
    print('D')
else:
    print('Faild')
```

```python
# Nested If Example
# if condition1:
#     if condition2:
#         if condition3:
```

```python
age=int(input("Enter Age: "))
if (age<13):
    print("not a teenager")
else:
    if (age<21):
        print("not eligible for marriage")
    else:
        print("eligible for marriage")
```

```python
# print 5 numbers using for loop
for temp in range(5): # condition is always n-1 
    print(temp)
```

```python
# print 1 to 5 numbers using for loop
for iam in range(1,5+1):
    print(iam,end=' ')  # automaticall \n is excued at end of print()
```

```python
# print numbers from  20 to 10 using for loop
for j in range(20,9,-1):
    print(j)
```

```python
# print even numbers b/w 2 dynamic integers
a,b = int(input("Enter A value: ")),int(input("Enter B Value: "))
for i in range(a,b+1):
    if i%2==0:
        print(i,end= '\t')
```

## Jumping Statments
- break
- continue
- pass

```python
# print numbers using dynamic integer with range of 1 to 10 using break
n =int(input("Enter N value= "))
for i in range(1,5+5+1):
    if i==n+1:
        break  # when break is executed loop will terminated
    print(i,end=",")
```

```python
# continue Example 
# print even Numbers from  2 dynamic integers using continue
x=int(input("Enter x: "));y=int(input('Enter Y:'))
for mn in range(x,y+1):
    if mn%2!=0:
        continue # skip the Iteration
    else:
        print(mn,end=" ")
```
